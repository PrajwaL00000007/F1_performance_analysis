```python
df = df.drop(columns=['Fastest Lap'])
```

```python
pd.isnull(df)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['Starting Grid'] = df['Starting Grid'].astype('int')
```

```python
df['Points'] = df['Points'].astype('int')
```

```python
ax = sns.countplot(x = 'Position' ,data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
win_position = df.groupby(['Driver'], as_index=False)['Points'].sum().sort_values(by='Points', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = win_position ,x =  'Driver',y = 'Points')
```

```python
win_position = df.groupby(['Driver'], as_index=False)['Starting Grid'].sum().sort_values(by='Starting Grid', ascending=True).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = win_position ,x =  'Driver',y = 'Starting Grid')
```

```python
win_position = df.groupby(['Team'], as_index=False)['Points'].sum().sort_values(by='Points', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = win_position ,x =  'Team',y = 'Points')
```

```python
win_position = df.groupby(['Driver'], as_index=False)['Points'].sum().sort_values(by='Points', ascending=False).head(10)
print((win_position.values.tolist()[0][0]))
```
